Exercicio 1

```python

numero = int(input("Digite um número inteiro: "))

if numero % 3 == 0 and numero % 5 == 0:
    print("O número é divisível por 3 e por 5 ao mesmo tempo.")
else:
    print("O número NÃO é divisível por 3 e por 5 ao mesmo tempo.")
```

Exercicio 2

```python

valor = float(input("Digite o valor da compra"))

if valor <= 1000:
    desconto = 0.10
elif valor <= 5000:
    desconto = 0.20
else:
    desconto = 0.30

valor_desconto = valor * desconto
valor_final = valor - valor_desconto


print(f"\nValor da compra: R$ {valor:.2f}")
print(f"Desconto aplicado: {desconto*100:.0f}%")
print(f"Valor do desconto: R$ {valor_desconto:.2f}")
print(f"Valor final a pagar: R$ {valor_final:.2f}")
```

Exercicio 3

```python
import math

largura = float(input("Digite a largura do aposento (em metros): "))
comprimento = float(input("Digite o comprimento do aposento (em metros): "))
litros_lata = float(input("Digite quantos litros tem uma lata de tinta: "))


area_total = largura * comprimento


area_porta = 0.80 * 2.10


area_pintura = area_total - area_porta


litros_necessarios = area_pintura / 3


latas_necessarias = math.ceil(litros_necessarios / litros_lata)


print(f"\nÁrea total: {area_total:.2f} m²")
print(f"Área a ser pintada (descontando a porta): {area_pintura:.2f} m²")
print(f"Litros de tinta necessários: {litros_necessarios:.2f} L")
print(f"Quantidade de latas necessárias: {latas_necessarias}")
```
